```python
import ClassiferHelperAPI as CH
import importlib
import pandas as pd
import csv
import json
importlib.reload(CH)
import numpy as np
import matplotlib.pyplot as plt
from sklearn import metrics
```

```python
# Remove unwanted columns
df = pd.DataFrame.from_csv("../FinalResults/ImgShrRnkListWithTags.csv")
cols = list(df.columns)
df.drop('URL',1,inplace=True)
df.drop('Album',1,inplace=True)
df.reset_index(inplace=True)
df = df.iloc[np.random.permutation(len(df))]
df.to_csv("/tmp/test.csv",index=False)
```

```python
reader = csv.reader(open("/tmp/test.csv","r"))
head = reader.__next__()
data = {}
for row in reader:
    temp = {}
    for i in range(1,len(row)):
        temp[head[i]] = row[i] 
    data[row[0]] = temp
    
allAttribs = CH.genHead(data,'SPECIES') + CH.genHead(data,'SEX') + CH.genHead(data,'AGE') + CH.genHead(data,'QUALITY') + CH.genHead(data,'VIEW_POINT') + CH.genHead(data,'INDIVIDUAL_NAME') + CH.genHead(data,'tags')
len(allAttribs)
```

```python
#CH.buildBinClassifier(data,allAttribs,0.5)

aucDict = {}
for i in np.arange(0.1,1,0.1):
    test_x, test_y, _, pred_prob = CH.buildBinClassifier(data,allAttribs,i,80,'random_forests')
    fpr, tpr, _ = metrics.roc_curve(test_y, pred_prob)
    auc = metrics.auc(fpr,tpr)
    title = str("AUC for train-test split of %f percent = %f" %(i*100,auc))
    aucDict[int(i*100)] = auc
    print(title)
```

```python
def getAccuracyDict(thres):    
    importlib.reload(CH)
    accuracy_dict  = {}
    for i in np.arange(0.1,1,0.1):
        temp = {}
        temp["Linear"] = CH.buildRegrModel(data,allAttribs,i,thres,False)
        temp["Logistic"] = CH.buildRegrModel(data,allAttribs,i,thres)

        accuracy_dict[i*100] = temp
        
    return accuracy_dict
```
